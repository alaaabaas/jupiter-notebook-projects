### Research Question 1:   what month sells best? !

```python
year_data['price_order']=year_data['Quantity Ordered']*year_data['Price Each']
```

### Research Question 2 :which city sells most?

```python
year_data.head()
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]

year_data['city']=year_data['Purchase Address'].apply(lambda x: get_city(x) +' '+ get_state(x))
```

```python
year_data
```

```python
city_sales=year_data.groupby('city').sum()
city_sales
```

```python
import matplotlib.pyplot as plt
cities=[city for city,df in year_data.groupby('city')]
plt.bar(cities,city_sales['price_order'])
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel('sales per city')

plt.show()
```

it seems that most of our customers are from san francisco

### Research Question 2 :which products are often sold together?

```python

year_data.head()
```

```python
 df=year_data[year_data['Order ID'].duplicated(keep=False)]
df
```

```python
   
df['together']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df
```

```python
df2=df.drop_duplicates()
```

```python
no_of_combinations=df2.groupby('together').count()
```

```python
df3=no_of_combinations.sort_values(by='Product',ascending=False)
```
