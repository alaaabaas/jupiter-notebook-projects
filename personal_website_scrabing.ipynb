```python
import pandas as pd
import re
from bs4 import BeautifulSoup as bs
import requests
```

```python
r=requests.get('https://keithgalli.github.io/web-scraping/webpage.html')
#convert to beautiful soup object
soup=bs(r.content)
#print out html
print(soup.prettify())
```

scarping the baseball table on the page

```python
table=soup.body.find('table')
```

```python
table
```

```python
header=table.find_all('th')
```

```python
header
```

```python
columns=[c.string for c in header]
```

```python
columns
```

we can see from the code that all the columns after '|' are empty so we can remove them from the begining

```python
rows=table.find('tbody').find_all('tr')
```

```python
rows
```

```python
l=[]
for tr in rows:
    row_data=tr.find_all('td')#get all cells data in arow tr
    row=[str(tr.get_text()).strip() for tr in row_data]#for each row get text strip to remove'\n'
    l.append(row)
```

tr:data in each row    td:data in each cell of the row 


```python
l
```

```python
baseball=pd.DataFrame(l,columns=columns)
```

```python
baseball
```

grabing all the fun facts that use the word is

```python
fun_facts=soup.body.select('ul.fun-facts li')
```

```python
fun_facts
```

```python
fun_is=[fun.find(string=re.compile("is")) for fun in fun_facts]
```

```python
fun_is
```

```python
fun_is=[fun for fun in fun_is if fun]
```

```python
fun_is
```

the sentences are not correct cropped after the word is to get the full sentence

```python
fun_is_neat=[fun.find_parent().get_text() for fun in fun_is]
```

```python
fun_is_neat
```
